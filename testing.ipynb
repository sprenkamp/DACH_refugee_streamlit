```python
import tiktoken
def num_tokens_from_string(string: str, encoding_name: str) -> int:
    """Returns the number of tokens in a text string."""
    encoding = tiktoken.get_encoding(encoding_name)
    num_tokens = len(encoding.encode(string))
    return num_tokens
```

```python
from tiktoken import Tokenizer, TokenCount
from tiktoken.tokenizer import Tokenizer

def count_tokens(text):
    tokenizer = Tokenizer()
    token_count = len(list(tokenizer.tokenize(text)))
    return token_count

sample_text = "This is an example sentence to count tokens."
tokens = count_tokens(sample_text)
print(f"Token count: {tokens}")
```
